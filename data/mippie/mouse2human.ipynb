```python
import numpy as np
import pandas as pd
from collections import defaultdict
```

```python
H = pd.read_csv('El_primo/DataFrame_humanGenes_matched-mouse_3col.csv')
H.head()
```

```python
hum = H[~H['ENTREZID'].isnull()].reset_index(drop=True)
hum['ENTREZID'] = hum['ENTREZID'].astype(int)
hum['SYMBOL'] = hum['SYMBOL'].str.upper()
hum = hum.drop(['GENEID'], axis=1)
print(len(hum.index))
hum.head()
```

```python
M = pd.read_csv('El_primo/mippie_proteins_v1_0.tsv', sep="\t")
M.head()
```

```python
M['SYMBOL'] = M['official_symbol'].str.upper()
mus = M.drop(['official_symbol','uniprot_accession','mgi'], axis=1)
print(len(mus.index))
mus.head()
```

```python
hum_dict = defaultdict(int)
hum_dict = {row[1]:row[0] for _,row in hum.iterrows()}
hum_dict
```

```python
hum_genes = []
for _, row in mus.iterrows():
    try:
      hum_genes.append(hum_dict[row[1]])
    except KeyError:
      hum_genes.append(None)
hum_genes
```

```python
type(hum_dict['ENO1'])
```

```python
mus['Human'] = hum_genes
DF = mus[~mus['Human'].isnull()]
DF['Human'] = DF['Human'].astype(int)
DF = DF.drop(['SYMBOL'], axis=1)
print(len(DF.index))
DF.head()
```

```python
print(len(DF['entrez'].unique()))
print(len(DF['SYMBOL'].unique()))
print(len(DF['Human'].unique()))
```

```python
DF.to_csv('./mouse2hum.csv',index=False,sep=',')
```
